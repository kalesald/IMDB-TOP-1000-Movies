```python
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
```

```python
df=pd.read_csv(r"C:\Users\Lenovo\Desktop\Data analysis\imdb_top_1000.csv")
```

```python
df
```

```python
df.drop(columns=['Poster_Link','Overview'],inplace=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df[df.duplicated()]
```

```python
df.dtypes
```

```python
df['Meta_score'] = df['Meta_score'].fillna(df['Meta_score'].median())
```

```python
df['Gross']=df['Gross'].str.replace(',','')
```

```python
df['Gross']=df['Gross'].fillna(0)
```

```python
df['Gross']=df['Gross'].astype(float)
```

```python
df['Gross']=df['Gross'].replace([0,0.0],df['Gross'].mean())
```

```python
df['Certificate']=df['Certificate'].fillna('UA')
```

```python
df.isna().sum()
```

```python
df['Runtime']=df['Runtime'].str.replace('min','')
```

```python
df['Runtime']=df['Runtime'].astype(int)
```

```python
df.info()
```

```python
df
```

```python
df['Certificate'].value_counts().plot(kind='barh', figsize=(10, 6))
plt.title("Certificate Count by type")
plt.ylabel("Certificates")
plt.xlabel("Count")
```

```python
genre_rating = df.groupby('Genre')['Meta_score'].mean().sort_values(ascending=False).head(10)

# Plot the average IMDb rating by Genre
genre_rating.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Average Meta Rating by Genre')
plt.ylabel('Average Meta Rating')
plt.xlabel('Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
genre_rating = df.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(10)

# Plot the average IMDb rating by Genre
genre_rating.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Average IMDb Rating by Genre')
plt.ylabel('Average IMDb Rating')
plt.xlabel('Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
top_10_grossing_movies = df.nlargest(10, 'Gross')
print("Top 10 Grossing Movies:")
print(top_10_grossing_movies[['Series_Title', 'Gross', 'Director']])
top_10_grossing_movies.plot(kind='barh', x='Series_Title', y='Gross', figsize=(15, 8), color='skyblue')
plt.title('Top 10 Most Grossing Movies')
plt.ylabel('Movie Title')
plt.xlabel('Gross Earning')
```

```python
director_ranking=df['Director'].value_counts().head(10)
print(director_ranking)
director_ranking.plot(kind="barh",figsize=(10,6),color="cyan")
plt.title("Top 10 Directors with most movies Directed")
plt.ylabel("Director")
plt.xlabel("Movies")
```

```python
director_gross=df.groupby('Director')['Gross'].sum().sort_values(ascending=False).head(10)
print(director_gross)
director_gross.plot(kind="barh",figsize=(10,6),color="red")
plt.title("Top 10 Directors by Gross Revenue")
plt.xlabel("Gross Revenue")
plt.ylabel("Director")
```

```python
all_stars = pd.concat([df['Star1'], df['Star2'], df['Star3'], df['Star4']])
all_actors = all_stars.value_counts().head(10)
print(all_actors)
all_actors.plot(kind='barh',figsize=(10,6),color='pink')
plt.title("Stars with the most apperance in movies")
plt.xlabel("Movies")
plt.ylabel("Stars")
```

```python
stars_df = df[['Star1', 'Star2', 'Star3', 'Star4']].melt(value_name='Star')
stars_gross = stars_df.merge(df[['Gross']], left_index=True, right_index=True)
stars_revenue = stars_gross.groupby('Star')['Gross'].sum().sort_values(ascending=False).head(10)
stars_revenue.plot(kind='barh', figsize=(10, 6), color='violet', title='Top 10 Stars by Gross Revenue')
plt.ylabel('Star')
plt.xlabel('Gross Revenue')
print(stars_revenue)
```

```python
df.groupby('Released_Year')["Runtime"].mean().plot(kind='line',figsize=(10,6),color='orange')
plt.title("Avereage Runtime over the years")
plt.xlabel('Year')
plt.ylabel('Runtime(minutes)')
```

```python
correlation_matrix = df[['IMDB_Rating','Gross','Meta_score','Runtime',"No_of_Votes"]].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
